```python
import pandas as pd
import numpy as np
```

```python
gender_df = pd.read_csv("gender_submission.csv")
test_df = pd.read_csv("test.csv")
train_df = pd.read_csv("train.csv")
```

```python
def clean_data(df):

  # Remove duplicates
  df.drop_duplicates(inplace = True)

  # fill missing numeric values
  numeric_cols = df.select_dtypes(include = ['float64','int64']).columns

  for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

  # fill missing categorical values
  categorical_cols = df.select_dtypes(include = ['object']).columns

  for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

  return df
```

```python
gender_cleaned = clean_data(gender_df)
test_cleaned = clean_data(test_df)
train_cleaned = clean_data(train_df)
```

```python
# checking missing values

print(train_cleaned.isnull().sum())
print(test_cleaned.isnull().sum())
print(gender_cleaned.isnull().sum())
```

```python
## Handling Missing values
# numerical values wereb filled using median values.
# categorical values were filled using mode values.
```

```python
print(train_cleaned.dtypes)
print(test_cleaned.dtypes)
print(gender_cleaned.dtypes)
```

```python
categorical_cols = ['Sex', 'Embarked', 'Cabin']

for col in categorical_cols:
    if col in train_cleaned.columns:
        train_cleaned[col] = train_cleaned[col].astype('category')
    if col in test_cleaned.columns:
        test_cleaned[col] = test_cleaned[col].astype('category')
```

```python
test_cleaned['Age'] = test_cleaned['Age'].round().astype('int64')
train_cleaned['Age'] = train_cleaned['Age'].round().astype('int64')
```

```python
print(train_cleaned.dtypes)
print(test_cleaned.dtypes)
```

```python
# save the files

gender_cleaned.to_csv("gender_submission_cleaned.csv", index=False)
test_cleaned.to_csv("test_cleaned.csv", index=False)
train_cleaned.to_csv("train_cleaned.csv", index=False)
```
